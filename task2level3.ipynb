```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\ragha\\Downloads\\Dataset .csv")
```

```python
display(df.head())
```

```python
cuisine_rating = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
print("Average rating for each cuisine type:")
print(cuisine_rating)
```

```python
top_cuisine_ratings = cuisine_rating.head(10)
top_cuisine_ratings.plot(kind='bar',title='Top cuisines by Average Rating')
```

```python
cuisine_votes = df.groupby('Cuisines')['Votes'].sum().sort_values(ascending=False)
print("Most popular cuisines based on the number of votes:")
print(cuisine_votes)
```

```python
top_cuisine_votes = cuisine_votes.head(10)  # Top 10 cuisines by votes
top_cuisine_votes.plot(kind='bar', title='Most Popular Cuisines by Votes')
```

```python
cuisine_popularity_rating = pd.DataFrame({
    'Average Rating': cuisine_rating,
    'Total Votes': cuisine_votes
})

# Display cuisines with high ratings and high vote counts
print("Cuisines with high ratings and high popularity:")
print(cuisine_popularity_rating.sort_values(by=['Average Rating', 'Total Votes'], ascending=False).head(10))
```
